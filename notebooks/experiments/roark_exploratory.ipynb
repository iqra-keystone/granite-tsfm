```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('C:\\Users\\IqraImtiaz\\OneDrive - keystonestrategy.com\\Documents\\Roark-Subway\\rest_only_transactions000000000000.csv.gz', parse_dates=['TRANSACTION_DATE'])
df
```

```python
df['ZIPCODE'].unique()
df = df.sort_values(by='TRANSACTION_DATE')
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['TRANSACTION_DATE'], df['GROSS_TRANSACTION_AMOUNT'])
plt.title('Transaction Date vs Gross Transaction Amount')
plt.xlabel('Transaction Date')
plt.ylabel('Gross Transaction Amount')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
